## Laboratory Work No. 1: Unsupervised Learning

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_moons, make_circles, make_classification, make_blobs
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
SAMPLE_SIZE = 1000
NOISE_LEVEL = 0.05
RANDOM_STATE = 42
```

```python
moon_data = make_moons(n_samples=SAMPLE_SIZE, noise=NOISE_LEVEL, random_state=RANDOM_STATE)
df_moon = pd.DataFrame(moon_data[0], columns=['Feature 1', 'Feature 2'])
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_moon, x='Feature 1', y='Feature 2')
plt.title('Moon-shaped Data')
plt.show()
```

```python
circle_data = make_circles(n_samples=SAMPLE_SIZE, noise=NOISE_LEVEL, factor=0.5, random_state=RANDOM_STATE)
df_circle = pd.DataFrame(circle_data[0], columns=['Feature 1', 'Feature 2'])
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_circle, x='Feature 1', y='Feature 2')
plt.title('Circle-shaped Data')
plt.show()
```

```python
cluster_data = make_classification(n_samples=SAMPLE_SIZE, n_features=2, n_informative=2, n_redundant=0,
                                   n_clusters_per_class=1, random_state=RANDOM_STATE)
df_cluster = pd.DataFrame(cluster_data[0], columns=['Feature 1', 'Feature 2'])
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_cluster, x='Feature 1', y='Feature 2')
plt.title('Cluster-shaped Data')
plt.show()
```

```python
blobs_data = make_blobs(n_samples=SAMPLE_SIZE, centers=3, cluster_std=1.0, random_state=RANDOM_STATE)
df_blobs = pd.DataFrame(blobs_data[0], columns=['Feature 1', 'Feature 2'])
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_blobs, x='Feature 1', y='Feature 2')
plt.title('Blob-shaped Data')
plt.show()
```

### K-Means Clustering

```python
moon_model_kmeans = KMeans(n_clusters=2, random_state=RANDOM_STATE)
moon_labels_kmeans = moon_model_kmeans.fit_predict(df_moon)
moon_score_kmeans = silhouette_score(df_moon, moon_labels_kmeans)
print(f"KMeans Silhouette Score on Moon-shaped Data: {moon_score_kmeans:.4f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_moon['Feature 1'], y=df_moon['Feature 2'], hue=moon_labels_kmeans, palette='Set1')
plt.title('KMeans Clustering on Moon-shaped Data')
plt.show()
```

```python
circle_model_kmeans = KMeans(n_clusters=2, random_state=RANDOM_STATE)
circle_labels_kmeans = circle_model_kmeans.fit_predict(df_circle)
circle_score_kmeans = silhouette_score(df_circle, circle_labels_kmeans)
print(f"KMeans Silhouette Score on Circle-shaped Data: {circle_score_kmeans:.4f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_circle['Feature 1'], y=df_circle['Feature 2'], hue=circle_labels_kmeans, palette='Set1')
plt.title('KMeans Clustering on Circle-shaped Data')
plt.show()
```

```python
cluster_model_kmeans = KMeans(n_clusters=2, random_state=RANDOM_STATE)
cluster_labels_kmeans = cluster_model_kmeans.fit_predict(df_cluster)
cluster_score_kmeans = silhouette_score(df_cluster, cluster_labels_kmeans)
print(f"KMeans Silhouette Score on Cluster-shaped Data: {cluster_score_kmeans:.4f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_cluster['Feature 1'], y=df_cluster['Feature 2'], hue=cluster_labels_kmeans, palette='Set1')
plt.title('KMeans Clustering on Cluster-shaped Data')
plt.show()
```

```python
blobs_model_kmeans = KMeans(n_clusters=3, random_state=RANDOM_STATE)
blobs_labels_kmeans = blobs_model_kmeans.fit_predict(df_blobs)
blobs_score_kmeans = silhouette_score(df_blobs, blobs_labels_kmeans)
print(f"KMeans Silhouette Score on Blob-shaped Data: {blobs_score_kmeans:.4f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_blobs['Feature 1'], y=df_blobs['Feature 2'], hue=blobs_labels_kmeans, palette='Set1')
plt.title('KMeans Clustering on Blob-shaped Data')
plt.show()
```

### Agglomerative Clustering

```python
moon_model_agglomerative = AgglomerativeClustering(n_clusters=2)
moon_labels_agglomerative = moon_model_agglomerative.fit_predict(df_moon)
moon_score_agglomerative = silhouette_score(df_moon, moon_labels_agglomerative)
print(f"Agglomerative Clustering Silhouette Score on Moon-shaped Data: {moon_score_agglomerative:.4f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_moon['Feature 1'], y=df_moon['Feature 2'], hue=moon_labels_agglomerative, palette='Set1')
plt.title('Agglomerative Clustering on Moon-shaped Data')
plt.show()
```

```python
circle_model_agglomerative = AgglomerativeClustering(n_clusters=2)
circle_labels_agglomerative = circle_model_agglomerative.fit_predict(df_circle)
circle_score_agglomerative = silhouette_score(df_circle, circle_labels_agglomerative)
print(f"Agglomerative Clustering Silhouette Score on Circle-shaped Data: {circle_score_agglomerative:.4f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_circle['Feature 1'], y=df_circle['Feature 2'], hue=circle_labels_agglomerative, palette='Set1')
plt.title('Agglomerative Clustering on Circle-shaped Data')
plt.show()
```

```python
cluster_model_agglomerative = AgglomerativeClustering(n_clusters=2)
cluster_labels_agglomerative = cluster_model_agglomerative.fit_predict(df_cluster)
cluster_score_agglomerative = silhouette_score(df_cluster, cluster_labels_agglomerative)
print(f"Agglomerative Clustering Silhouette Score on Cluster-shaped Data: {cluster_score_agglomerative:.4f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_cluster['Feature 1'], y=df_cluster['Feature 2'], hue=cluster_labels_agglomerative, palette='Set1')
plt.title('Agglomerative Clustering on Cluster-shaped Data')
plt.show()
```

```python
blob_model_agglomerative = AgglomerativeClustering(n_clusters=3)
blob_labels_agglomerative = blob_model_agglomerative.fit_predict(df_blobs)
blob_score_agglomerative = silhouette_score(df_blobs, blob_labels_agglomerative)
print(f"Agglomerative Clustering Silhouette Score on Blob-shaped Data: {blob_score_agglomerative:.4f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_blobs['Feature 1'], y=df_blobs['Feature 2'], hue=blob_labels_agglomerative, palette='Set1')
plt.title('Agglomerative Clustering on Blob-shaped Data')
plt.show()
```

### DBSCAN

```python
circle_model_dbscan = DBSCAN(eps=0.2, min_samples=5)
circle_labels_dbscan = circle_model_dbscan.fit_predict(df_circle)
circle_score_dbscan = silhouette_score(df_circle, circle_labels_dbscan)
print(f"DBSCAN Silhouette Score on Circle-shaped Data: {circle_score_dbscan:.4f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_circle['Feature 1'], y=df_circle['Feature 2'], hue=circle_labels_dbscan, palette='Set1')
plt.title('DBSCAN Clustering on Circle-shaped Data')
plt.show()
```

```python
cluster_model_dbscan = DBSCAN(eps=0.5, min_samples=5)
cluster_labels_dbscan = cluster_model_dbscan.fit_predict(df_cluster)
cluster_score_dbscan = silhouette_score(df_cluster, cluster_labels_dbscan)
print(f"DBSCAN Silhouette Score on Cluster-shaped Data: {cluster_score_dbscan:.4f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_cluster['Feature 1'], y=df_cluster['Feature 2'], hue=cluster_labels_dbscan, palette='Set1')
plt.title('DBSCAN Clustering on Cluster-shaped Data')
plt.show()
```

```python
blobs_model_dbscan = DBSCAN(eps=0.5, min_samples=5)
blobs_labels_dbscan = blobs_model_dbscan.fit_predict(df_blobs)
blobs_score_dbscan = silhouette_score(df_blobs, blobs_labels_dbscan)
print(f"DBSCAN Silhouette Score on Blob-shaped Data: {blobs_score_dbscan:.4f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_blobs['Feature 1'], y=df_blobs['Feature 2'], hue=blobs_labels_dbscan, palette='Set1')
plt.title('DBSCAN Clustering on Blob-shaped Data')
plt.show()
```
