## Laboratory Work No. 2: Supervised Learning

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, ConfusionMatrixDisplay
from sklearn.datasets import load_digits

from matplotlib import pyplot as plt
```

```python
dataset = load_digits(as_frame=True)
x_train, x_test, y_train, y_test = train_test_split(dataset.data, dataset.target, test_size=0.2, random_state=42)
```

```python
ds_df = dataset.frame
ds_df.head()
```

### Gaussian Naive Bayes Classifier

```python
gaussian_model = GaussianNB()
gaussian_model.fit(x_train, y_train)
gaussian_predictions = gaussian_model.predict(x_test)

print("Gaussian Naive Bayes Classifier Results:")
print(classification_report(y_test, gaussian_predictions))
print("Accuracy:", accuracy_score(y_test, gaussian_predictions))

ConfusionMatrixDisplay(confusion_matrix(y_test, gaussian_predictions)).plot()
```

### Support Vector Machine Classifier

```python
svc_model = SVC()
svc_model.fit(x_train, y_train)
svc_predictions = svc_model.predict(x_test)
print("Support Vector Machine Classifier Results:")
print(classification_report(y_test, svc_predictions))
print("Accuracy:", accuracy_score(y_test, svc_predictions))

ConfusionMatrixDisplay(confusion_matrix(y_test, svc_predictions)).plot()
```

### Decision Tree Classifier

```python
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(x_train, y_train)
decision_tree_predictions = decision_tree_model.predict(x_test)

print("Decision Tree Classifier Results:")
print(classification_report(y_test, decision_tree_predictions))
print("Accuracy:", accuracy_score(y_test, decision_tree_predictions))

ConfusionMatrixDisplay(confusion_matrix(y_test, decision_tree_predictions)).plot()
```

### AdaBoost Classifier

```python
ada_boost_model = AdaBoostClassifier()
ada_boost_model.fit(x_train, y_train)
ada_boost_predictions = ada_boost_model.predict(x_test)

print("AdaBoost Classifier Results:")
print(classification_report(y_test, ada_boost_predictions))
print("Accuracy:", accuracy_score(y_test, ada_boost_predictions))

ConfusionMatrixDisplay(confusion_matrix(y_test, ada_boost_predictions)).plot()
```
